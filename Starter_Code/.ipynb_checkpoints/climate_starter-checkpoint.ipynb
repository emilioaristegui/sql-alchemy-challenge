# Reflect Tables into SQLAlchemy ORM

```python
# Prepares the foundation for interacting with a database using SQLAlchemy.

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy

# .automap class from module allows automatic mapping of database tables to Python classes.
from sqlalchemy.ext.automap import automap_base

# .orm class from module to interact with the database through Python objects.
from sqlalchemy.orm import Session

# create_engine function to connect to database and func to provide SQL functions for SQLAlchemy queries. 
from sqlalchemy import create_engine, func
```

```python
# Engine used to perform various database operations: executing SQL queries, interacting with the database tables
# and using SQLAlchemy's ORM or SQL expressions.

# create engine to hawaii.sqlite
# create_engine function creates a database engine object that acts as a interface to the database.
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# Create a set of Python classes generated by SQLAlchemy that represent the tables in the reflected database schema.
# Used to query, insert, update, and delete data in the database using SQLAlchemy's ORM capabilities.

# reflect an existing database into a new model
# automap_base() function creates a base class to work as foundation for Python classes from the tables in the db.
Base = automap_base()

# reflect the tables
# prepare() method on the Base class, an instance of automap_base, reflects the tables in the db and generates Python classes.
# engine object, represents the db connection to be used SQLAlchemy knows which db to connect and retrieve information.
Base.prepare(autoload_with=engine)
```

```python
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Access and work with the corresponding Python classes representing those tables. 

# Save references to each table
# store in variables
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
# To see the names of the columns in the Measurement table, useful for understanding the structure of the table
# and referencing specific columns when working with data from the table.

# inspect function retrieves information about database tables: column names, data types, and constraints.
from sqlalchemy.inspection import inspect

# inspect function takes a table or a class mapped to a table, returns an inspector object with table's info structure.
table = inspect(Measurement)

# Iterates over columns of table and table.c attribute provides access to the columns of the table.
for column in table.c:
    
    # column.name attribute returns the name of the column.
    print(column.name)
```

```python
# inspect function takes a table or a class mapped to a table, returns an inspector object with table's info structure.
table = inspect(Station)

# Iterates over columns of table and table.c attribute provides access to the columns of the table.
for column in table.c:
    
    # column.name attribute returns the name of the column.
    print(column.name)
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
# in sql, SELECT max(DATE) FROM Measurement
# this is the ORM code
max_date = session.query(
    func.max(Measurement.date)).all()
max_date[0]
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 

# Calculate the date one year from the last date in data set.
# dt.date() creates a date object
# dt.timedelta() creates a timedelta object
# subtraction results in a new date object stored in query_date
query_date = dt.date(2017, 8, 23) - dt.timedelta(days=365)
print("Query Date: ", query_date)

# Perform a query to retrieve the data and precipitation scores
# session.query () method to select the date and prcp columns
prcp_query = session.query(Measurement.date, Measurement.prcp).\
# filter() specifies that date column table should be >= to query_date to include only records with dates on the range.
filter(Measurement.date >= query_date).\
    order_by(Measurement.date).all()

# Save the query results as a Pandas DataFrame. Explicitly set the column names
prcp_df = pd.DataFrame(prcp_query, columns=["date","prcp"])

# Sort the dataframe by date
prcp_df = prcp_df.sort_values("date")
prcp_df = prcp_df.set_index("date")

# Use Pandas Plotting with Matplotlib to plot the data
prcp_df.plot()
plt.xticks(rotation=90)
plt.show()
```

```python
# Use Pandas to calculate the summary statistics for the precipitation data
prcp_df.describe()
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number of stations in the dataset
# func.count() function is an SQL function from SQLAlchemy to perform a count operation.
# scalar() is called on the query object to execute the query and retrieve the scalar result (the count of stations).
total_stations = session.query(func.count(Station.station)).scalar()

print("Total number of stations: ", total_stations)
```

```python
# Design a query to find the most active stations (i.e. which stations have the most rows?)
# func.count() performs a count operation
active_stations = session.query(Measurement.station, func.count(Measurement.station)).\
    # group_by() groups the rows by the station column.
    group_by(Measurement.station).\
    # List the stations and their counts in descending order.
    # order_by() orders the groups in descending order based on the count of rows in each station.
    # .all() executes the query and retrieves all the matching records as a list of tuples
    order_by(func.count(Measurement.station).desc()).all()

for station, row_count in active_stations:
    print(f"Station: {station}, Row Count: {row_count}")
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
# Get the most active station based on row count
most_active_station = active_stations[0][0]  # Assuming active_stations is the result from the previous query

# Perform a query to calculate the lowest, highest, and average temperature for the most active station
temperature_stats = session.query(func.min(Measurement.tobs), func.max(Measurement.tobs), func.avg(Measurement.tobs)).\
    # filter() filters the data to include only measurements from the most active station.
    # .all() executes the query and retrieves all the matching records as a list of tuples.
    filter(Measurement.station == most_active_station).all()

# Extract the temperature statistics
lowest_temp, highest_temp, avg_temp = temperature_stats[0]

# Print the temperature statistics
print(f"Lowest Temperature: {lowest_temp}")
print(f"Highest Temperature: {highest_temp}")
print(f"Average Temperature: {avg_temp}")
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram

# Perform a query to retrieve the last 12 months of temperature observation data for the most active station
temperature_data = session.query(Measurement.tobs).\
    filter(Measurement.station == most_active_station).\
    filter(Measurement.date >= query_date).all()

# Extract the temperatures from the query result
temperatures = [result[0] for result in temperature_data]

# Plot the histogram
plt.hist(temperatures, bins=12, edgecolor='black')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.title('Temperature Observation Data')
plt.show()
```
